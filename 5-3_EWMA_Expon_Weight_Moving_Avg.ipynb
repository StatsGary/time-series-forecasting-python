```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
airline = pd.read_csv('Data/airline_passengers.csv', index_col='Month')
```

```python
airline.dropna(inplace=True)
airline.index = pd.to_datetime(airline.index)
```

```python
airline.head()
```

```python
airline['6_month_SMA'] = airline['Thousands of Passengers'].rolling(window=6).mean()
```

```python
airline['12_month_SMA'] = airline['Thousands of Passengers'].rolling(window=12).mean()
```

```python
airline.head()
```

```python
airline.plot(figsize=(12,6))
```

```python
# Exponentially weighted moving average
airline['EWMA-12'] = airline['Thousands of Passengers'].ewm(span=12).mean()
```

```python
airline[['Thousands of Passengers', 'EWMA-12']].plot(figsize=(10,6))
```
